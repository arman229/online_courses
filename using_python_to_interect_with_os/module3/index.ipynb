Regex
Regex (short for Regular Expression) is a special pattern used to search, match, or replace text. 
find all phone numbers
find all emails 
replace all dates with a new format


```python
# Simple matching in python
# r is raw string: A raw string is a string prefixed with 'r' or 'R' that tells Python to treat backslashes as literal characters and not as escape characters.
import re
resp = re.search(R'i', 'Hello i am arman ashraf,i live in lahore')
print(resp)
print(resp[0])
```

```python
# startwith(^:Start of line) and endwith($:End of line) in python regex
arman_rol = """Arman is a software engineer."""
resp = re.search(R'^Arman', arman_rol)
print(f'Matched : {resp}')
resp = re.search(R'^arman', arman_rol)
print(f"Not matched: {resp}")
resp = re.search(R'engineer.$', arman_rol)
print(f"matched: {resp}")
resp = re.search(R'engineerS$', arman_rol)
print(f"Not matched: {resp}")

```

```python
# .(dot) matches any one character except newline 
resp = re.search(R'a.b', 'anb')
print(f'Matched: {resp}')
resp = re.search(R'a.b', 'Anb',re.IGNORECASE) # ignore case
print(f'Matched: {resp}')
```

```python
# [] matches any one character inside the bracket
resp = re.search(R'[aeiou]', 'hello')
print(f'Matched: {resp}')
```

```python
# | matches either or condition
resp = re.search(r'cat|dog', 'i have a cat and a dog')
print(f'Matched: {resp}')
```

```python
# ^ inside bracket[] means negation
resp = re.search(r'[^aeiou]', 'ai is a vowel')
print(f'Matched: {resp}')
```

```python
# * matches zero or more times
resp = re.search(r'ab*c', 'ac')
print(f'Matched: {resp}')
resp = re.search(r'ab.*c', 'abc')
print(f'Matched: {resp}')
```

```python
# ? matches zero or one time
resp = re.search(r'ab?c', 'ac')
print(f'Matched: {resp}')
resp = re.search(r'ab?c', 'abdc')
print(f'Matched: {resp}')
```

```python
# + matches one or more times
resp = re.search(r'ab+c', 'ac')
print(f'Not Matched: {resp}')
resp = re.search(r'b+c+', 'abcccc')  # at least one time  
print(f'Matched: {resp}') 

resp = re.search(r'bc', 'abbcccc') # exactly one time
print(f'Matched: {resp}') 
```

```python
# \. matches literal dot
resp = re.search(r'a\.b', 'a.b')
print(f'Matched: {resp}')
resp = re.search(r'a\.b', 'acb')
print(f'Not Matched: {resp}')
```

```python
# \w matches any alphanumeric character (letters, digits, and underscores)
resp = re.search(r'\w+', 'Hello_123!')
print(f'Matched: {resp}')
```

```python
# \d matches any digit (0-9)
resp = re.search(r'\d+', 'My number is 12345')
print(f'Matched: {resp}')
```

```python
# () groups patterns together
resp = re.search(r'(ab)+', 'abababxyz')
print(f'Matched: {resp}')
```

```python
# splite using regex
resp = re.split(r'\s+', 'Hello   world  this is   regex')
print(f'Split Result: {resp}')


resp = re.split(r'[this]', 'Hello world this is regex')
print(f'Split Result: {resp}')


resp = re.split(r'([Hthis])', 'Hello   world  this is   regex')
print(f'Split Result: {resp}')
```
